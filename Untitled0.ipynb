```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# 1. 读取数据
# ===============================
creditcard_df = pd.read_csv("/content/creditcard.csv")

# ===============================
# 2. 单变量分布分析函数
# ===============================
def plot_univariate_distribution(
    data,
    column,
    bins=50,
    dataset_name="Dataset",
    save_path=None
):
    """
    论文级单变量分布分析图：
    上：直方图
    下：箱线图
    """

    values = data[column].dropna()

    fig, axes = plt.subplots(
        nrows=2,
        ncols=1,
        figsize=(7, 6),
        gridspec_kw={'height_ratios': [3, 1]}
    )

    # ---------- 1. Histogram ----------
    axes[0].hist(values, bins=bins)
    axes[0].set_title(
        f"Distribution of {column} ({dataset_name})",
        fontsize=12
    )
    axes[0].set_ylabel("Frequency")
    axes[0].grid(True, linestyle='--', alpha=0.5)

    # ---------- 2. Boxplot ----------
    axes[1].boxplot(
        values,
        vert=False,
        showfliers=True
    )
    axes[1].set_xlabel(column)
    axes[1].grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")

    plt.show()

# ===============================
# 3. 关键变量绘图（论文用）
# ===============================

# 交易金额 Amount
plot_univariate_distribution(
    data=creditcard_df,
    column="Amount",
    dataset_name="Credit Card Fraud",
    save_path="fig_3_1_amount_distribution.png"
)


# PCA 特征 V2（选一个代表性即可）
plot_univariate_distribution(
    data=creditcard_df,
    column="V2",
    dataset_name="Credit Card Fraud",
    save_path="fig_3_2_v2_distribution.png"
)
```

```python
# =========================
# Classical SVM Baseline
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, roc_auc_score

# ① Load data
df = pd.read_csv("/content/creditcard.csv")

X = df.drop(columns=["Class"])
y = df["Class"]

# ② Train / Test split
# 丢掉 y 是 NaN 的样本（同步 X）
mask = y.notna()
X = X[mask]
y = y[mask].astype(int)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# ③ Standardization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# ④ SVM (RBF)
svm = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale",
    class_weight="balanced",
    probability=True,
    random_state=42
)

svm.fit(X_train_scaled, y_train)

# ⑤ Accuracy
acc = svm.score(X_test_scaled, y_test)
print("SVM Accuracy:", acc)

# ⑥ Confusion Matrix
y_pred = svm.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix - SVM (RBF)")
plt.show()

# ⑦ ROC-AUC
y_score = svm.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_score)
auc = roc_auc_score(y_test, y_score)

print(f"SVM ROC-AUC: {auc:.4f}")

plt.figure()
plt.plot(fpr, tpr, label=f"SVM (AUC = {auc:.4f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - SVM (Classical Baseline)")
plt.legend()
plt.show()
```

```python
# ================================
# Credit Card Fraud - QSVM
# ================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc

from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel

# 1️⃣ 读取数据
df = pd.read_csv("/content/creditcard.csv")

df_0 = df[df["Class"] == 0]
df_1 = df[df["Class"] == 1]

n0 = min(100, len(df_0))
n1 = min(100, len(df_1))

df_0_small = df_0.sample(n=n0, random_state=42)
df_1_small = df_1.sample(n=n1, random_state=42)

df_q = pd.concat([df_0_small, df_1_small]).sample(frac=1, random_state=42)

print("Class 0:", n0, "Class 1:", n1)

X = df_q.drop(columns=["Class"])
y = df_q["Class"].values

# 2️⃣ 预处理
X = StandardScaler().fit_transform(X)
X = PCA(n_components=2).fit_transform(X)

# 3️⃣ 划分数据
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# 4️⃣ 量子核
feature_map = ZZFeatureMap(feature_dimension=2, reps=2)
qkernel = FidelityQuantumKernel(feature_map=feature_map)

# 5️⃣ QSVM（推荐：自动 kernel）
qsvm = SVC(kernel=qkernel.evaluate)
qsvm.fit(X_train, y_train)

# 6️⃣ 预测
y_pred = qsvm.predict(X_test)
y_score = qsvm.decision_function(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print("QSVM Accuracy:", acc)

# 7️⃣ 混淆矩阵
cm = confusion_matrix(y_test, y_pred)

# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")


# 8️⃣ ROC 曲线
fpr, tpr, _ = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ---------- 左：Confusion Matrix ----------
disp = ConfusionMatrixDisplay(cm)
disp.plot(
    ax=axes[0],
    cmap="viridis",
    colorbar=False
)
axes[0].set_title("Confusion Matrix - QSVM (Credit Card Fraud)")
axes[0].set_xlabel("Predicted Label")
axes[0].set_ylabel("True Label")

# ---------- 右：ROC Curve ----------
axes[1].plot(fpr, tpr, label=f"QSVM (AUC = {roc_auc:.4f})")
axes[1].plot([0, 1], [0, 1], "r--")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve - QSVM (Credit Card Fraud)")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# =========================================
# Credit Card Fraud Detection - NN Baseline
# =========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, roc_auc_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam

# 1️⃣ 读取数据
df = pd.read_csv("/content/creditcard.csv")

# 特征 & 标签
X = df.drop(columns=["Class"])
y = df["Class"]

# 只保留数值特征（保险）
X = X.select_dtypes(include=[np.number])

mask = y.notna()
X = X[mask]
y = y[mask].astype(int)

# print("Data shape:", X.shape)
# print("Fraud ratio:", y.mean())

# 2️⃣ 标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3️⃣ 训练 / 测试集划分（保持类别比例）
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# 4️⃣ 构建 NN（浅层，论文 baseline）
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(8, activation="relu"),
    Dense(4, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# 5️⃣ 训练
history = model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=256,
    validation_split=0.2,
    verbose=1
)

# 6️⃣ 预测
y_pred_prob = model.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

# 7️⃣ Accuracy
acc = np.mean(y_pred == y_test)
print(f"Credit Card Fraud - NN accuracy: {acc:.4f}")

# 8️⃣ Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")


# 9️⃣ ROC Curve & AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
auc = roc_auc_score(y_test, y_pred_prob)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ---------- 左：Confusion Matrix ----------
disp = ConfusionMatrixDisplay(cm)
disp.plot(
    ax=axes[0],
    cmap="viridis",
    colorbar=False
)
axes[0].set_title("Confusion Matrix - NN (Credit Card Fraud)")

# ---------- 右：ROC Curve ----------
axes[1].plot(fpr, tpr, label=f"NN (AUC = {auc:.4f})")
axes[1].plot([0, 1], [0, 1], "r--")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve - NN (Credit Card Fraud)")
axes[1].legend()

plt.tight_layout()
plt.show()


print(f"NN ROC-AUC: {auc:.4f}")
```

```python
# Bank Account Fraud（Basic） LR
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_auc_score
import matplotlib.pyplot as plt

# ① 读取数据
df_baf = pd.read_csv("/content/Base.csv")

# print(df_baf.shape)
# print(df_baf.columns)
# df_baf.head()

# ② 只保留数值列
# print(df_baf.columns.tolist())
y = df_baf["fraud_bool"]
X = df_baf.drop(columns=["fraud_bool"])
X = X.select_dtypes(include=["int64", "float64"])

# 标准化 + PCA + 归一化
# ① 标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ② inf / -inf → NaN（非常关键）
X_scaled = np.where(np.isfinite(X_scaled), X_scaled, np.nan)

# ③ 用中位数填补 NaN
imputer = SimpleImputer(strategy="median")
X_filled = imputer.fit_transform(X_scaled)

# ④ PCA → 2 维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_filled)

# ⑤ 映射到 (-1, 1)
mm = MinMaxScaler(feature_range=(-1, 1))
X_final = mm.fit_transform(X_pca)

# 划分数据 + LR
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.3, random_state=42, stratify=y
)

# 1. 训练模型
clf = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    random_state=42
)
clf.fit(X_train, y_train)

# 2. Accuracy
train_acc = clf.score(X_train, y_train)
test_acc = clf.score(X_test, y_test)
print(f"Train accuracy: {train_acc:.4f}")
print(f"Test accuracy : {test_acc:.4f}")

# 3. 混淆矩阵
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
ConfusionMatrixDisplay(cm).plot()
plt.show()

# 4. ROC-AUC
y_score = clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_score)
auc = roc_auc_score(y_test, y_score)
print(f"ROC-AUC: {auc:.4f}")
# 画 ROC 曲线
plt.figure()
plt.plot(fpr, tpr, label=f"LR (AUC = {auc:.4f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Account Fraud(Basic)(LR)")
plt.legend()
plt.show()
```

```python
# Bank Account Fraud（Basic）– SVM
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import roc_curve, roc_auc_score

# ① 读取数据
df = pd.read_csv("/content/Base.csv")
# print(df.shape)

# ② 特征 / 标签分离（非常关键）
y = df["fraud_bool"].astype(int)

X = df.drop(columns=["fraud_bool"])
X = X.select_dtypes(include=[np.number])  # 只保留数值特征

# ③ 处理 NaN / inf
# 把 inf / -inf 变成 NaN
X = X.replace([np.inf, -np.inf], np.nan)

# 用中位数填补
imputer = SimpleImputer(strategy="median")
X_filled = imputer.fit_transform(X)

# ④ 标准化（SVM 必须）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filled)

# ⑤ 划分训练 / 测试集（保持类别比例）
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# ⑥ 训练 SVM（RBF 核，论文级标准）
svm = LinearSVC(
    class_weight="balanced",
    max_iter=5000,
    random_state=42
)

svm.fit(X_train, y_train)

# ⑦ Accuracy
acc = svm.score(X_test, y_test)
print(f"Bank Account Fraud (Basic) - SVM accuracy: {acc:.4f}")

# 混淆矩阵
y_pred = svm.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")

# ROC 曲线 + AUC
y_score = svm.decision_function(X_test)

auc = roc_auc_score(y_test, y_score)
print(f"SVM ROC-AUC: {auc:.4f}")

fpr, tpr, _ = roc_curve(y_test, y_score)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ---------- 左：Confusion Matrix ----------
disp = ConfusionMatrixDisplay(cm)
disp.plot(
    ax=axes[0],
    cmap="viridis",
    colorbar=False
)
axes[0].set_title("Confusion Matrix - SVM (Bank Account Fraud Basic)")
axes[0].set_xlabel("Predicted Label")
axes[0].set_ylabel("True Label")

# ---------- 右：ROC Curve ----------
axes[1].plot(fpr, tpr, label=f"QSVM (AUC = {roc_auc:.4f})")
axes[1].plot([0, 1], [0, 1], "r--")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve - SVM (Classical Baseline)")
axes[1].legend()

plt.tight_layout()
plt.show()

```

```python
# ================================
# Bank Account Fraud (Basic) - QSVM
# ================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc

from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel

# ① 读取数据
df = pd.read_csv("/content/Base.csv")

# ② 只保留数值特征
X = df.drop(columns=["fraud_bool"])
X = X.select_dtypes(include=[np.number])
y = df["fraud_bool"].astype(int)

# ③ 查看类别数量（很重要）
print("Class counts:")
print(y.value_counts())

# ④ 强制小样本（QSVM 必须）
n_per_class = 100

df_0 = df[df["fraud_bool"] == 0].sample(
    n=min(n_per_class, (y == 0).sum()), random_state=42
)
df_1 = df[df["fraud_bool"] == 1].sample(
    n=min(n_per_class, (y == 1).sum()), random_state=42
)

df_q = pd.concat([df_0, df_1]).sample(frac=1, random_state=42)

X = df_q.drop(columns=["fraud_bool"])
X = X.select_dtypes(include=[np.number])
y = df_q["fraud_bool"].values

print("QSVM dataset shape:", X.shape)
print("Class 0:", (y == 0).sum(), "Class 1:", (y == 1).sum())

# ⑤ 标准化 + PCA → 2 维
X = StandardScaler().fit_transform(X)
X = PCA(n_components=2).fit_transform(X)

# ⑥ 划分训练 / 测试集
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# ⑦ 构造量子特征映射 & 量子核
feature_map = ZZFeatureMap(feature_dimension=2, reps=2)
quantum_kernel = FidelityQuantumKernel(feature_map=feature_map)

# ⑧ QSVM（注意：kernel 用 callable）
qsvm = SVC(kernel=quantum_kernel.evaluate, probability=True)
qsvm.fit(X_train, y_train)

# ⑨ 预测
y_pred = qsvm.predict(X_test)
y_score = qsvm.predict_proba(X_test)[:, 1]

# ⑩ Accuracy
acc = accuracy_score(y_test, y_pred)
print("QSVM Accuracy:", acc)

# ⑪ 混淆矩阵
cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")

# ⑫ ROC 曲线
fpr, tpr, _ = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)

plt.show()
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ---------- 左：Confusion Matrix ----------
disp = ConfusionMatrixDisplay(cm)
disp.plot(
    ax=axes[0],
    cmap="viridis",
    colorbar=False
)
axes[0].set_title("Confusion Matrix - QSVM (Bank Account Fraud)")
axes[0].set_xlabel("Predicted Label")
axes[0].set_ylabel("True Label")

# ---------- 右：ROC Curve ----------
axes[1].plot(fpr, tpr, label=f"QSVM (AUC = {roc_auc:.4f})")
axes[1].plot([0, 1], [0, 1], "r--")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve - QSVM (Bank Account Fraud)")
axes[1].legend()

plt.tight_layout()
```

```python
# ================================
# Bank Account Fraud (Basic) - NN
# ================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    roc_auc_score
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------------
# 1. Load data
# -----------------------------
df = pd.read_csv("/content/Base.csv")

y = df["fraud_bool"].astype(int).values
X = df.drop(columns=["fraud_bool"])
X = X.select_dtypes(include=[np.number])

print("Dataset shape:", X.shape)
print("Class distribution:", np.bincount(y))

# -----------------------------
# 2. Handle NaN / inf
# -----------------------------
X = X.replace([np.inf, -np.inf], np.nan)

imputer = SimpleImputer(strategy="median")
X_filled = imputer.fit_transform(X)

# -----------------------------
# 3. Standardization (NN 必须)
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filled)

# -----------------------------
# 4. Train / test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# -----------------------------
# 5. Build NN model
# -----------------------------
nn = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(16, activation="relu"),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# -----------------------------
# 6. Early stopping
# -----------------------------
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

# -----------------------------
# 7. Train
# -----------------------------
history = nn.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=128,
    callbacks=[early_stop],
    verbose=1
)

# -----------------------------
# 8. Prediction
# -----------------------------
y_pred_prob = nn.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

# -----------------------------
# 9. Metrics
# -----------------------------
cm = confusion_matrix(y_test, y_pred)
# ---------- 评价指标计算 ----------
tn, fp, fn, tp = cm.ravel()

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")

# -----------------------------
# 10. Confusion Matrix
# -----------------------------
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix - NN (Bank Account Fraud Basic)")
plt.show()

# -----------------------------
# 11. ROC - AUC
# -----------------------------
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
auc = roc_auc_score(y_test, y_pred_prob)

plt.figure()
plt.plot(fpr, tpr, label=f"NN (AUC = {auc:.4f})")
plt.plot([0, 1], [0, 1], "r--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - NN (Bank Account Fraud Basic)")
plt.legend()
plt.show()

print(f"NN ROC-AUC: {auc:.4f}")
```
